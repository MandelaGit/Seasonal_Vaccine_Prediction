## Prediction of Seasonal Flu Vaccinations

### Business Understanding

Designing successful vaccination plans in the context of seasonal vaccines requires a thorough awareness of the financial consequences and aspects. In order to maximize vaccination uptake and raise the acceptance of seasonal vaccines, this initiative aims to give stakeholders, including healthcare organizations, public health authorities, and legislators, with actionable knowledge.

In order to reach high immunization rates among target demographics, businesses and healthcare professionals must overcome many major obstacles. Organizations can modify their communication and outreach efforts to address potential obstacles and concerns by accurately forecasting vaccine uptake or refusal. This can involve creating carefully crafted marketing campaigns, engaging the community, and giving pertinent information to people who might be wary of vaccinations. Predicting vaccine adoption can also assist in planning resource allocation and distribution.

#### Problem Statement

The challenge is to create a prediction model that can anticipate people's acceptance or rejection of seasonal immunizations. For reducing the spread of infectious illnesses and ensuring the health of communities, vaccine adoption must increase. However, achieving high immunization rates is significantly hampered by vaccine reluctance and rejection. In order to inform targeted treatments and improve vaccination tactics, it is necessary to address the issue of accurately forecasting vaccine acceptance.

#### Main Objective

The objective of this investigation is to develop a model that can forecast a person's acceptance or refusal of the seasonal flu vaccine based on their history and behavioural characteristics.

#### Metric for Success
- The accuracy of the models will inform on their success or lack thereof. The ideal accuracy that the project is aiming for is 80%. 

#### Description/Summary of Dataset

The dataset was download from Driven Data as part of a project that aims at predicting how likely people are to receive the H1N1 and seasonal flu vaccines. The data was part of a 2009 National H1N1 Flu Survey and contains 35 features and 2 target variables. The sects of datasets are as showcased below.

- **For this project, the seasonal_vaccine column was used as the target variable because it did not have a lot of imbalance between its binary variables**

- **Target Variables**
1. `h1n1_vaccine` - Whether respondent received H1N1 flu vaccine.
2. `seasonal_vaccine` - Whether respondent received seasonal flu vaccine.
Both are binary variables: 0 = No; 1 = Yes.

- **Independent Features**
1. `h1n1_concern` - Level of concern about the H1N1 flu.
0 = Not at all concerned; 1 = Not very concerned; 2 = Somewhat concerned; 3 = Very concerned.
2. `h1n1_knowledge` - Level of knowledge about H1N1 flu.
0 = No knowledge; 1 = A little knowledge; 2 = A lot of knowledge.
3. `behavioral_antiviral_meds` - Has taken antiviral medications. (binary)
4. `behavioral_avoidance` - Has avoided close contact with others with flu-like symptoms. (binary)
5. `behavioral_face_mask` - Has bought a face mask. (binary)
6. `behavioral_wash_hands` - Has frequently washed hands or used hand sanitizer. (binary)
7. `behavioral_large_gatherings` - Has reduced time at large gatherings. (binary)
8. `behavioral_outside_home` - Has reduced contact with people outside of own household. (binary)
9. `behavioral_touch_face` - Has avoided touching eyes, nose, or mouth. (binary)
10. `doctor_recc_h1n1` - H1N1 flu vaccine was recommended by doctor. (binary)
11. `doctor_recc_seasonal` - Seasonal flu vaccine was recommended by doctor. (binary)
12. `chronic_med_condition` - Has any of the following chronic medical conditions: asthma or an other lung condition, diabetes, a heart condition, a kidney condition, sickle cell anemia or other anemia, a neurological or neuromuscular condition, a liver condition, or a weakened immune system caused by a chronic illness or by medicines taken for a chronic illness. (binary)
13. `child_under_6_months` - Has regular close contact with a child under the age of six months. (binary)
14. `health_worker` - Is a healthcare worker. (binary)
15. `health_insurance` - Has health insurance. (binary)
16. `opinion_h1n1_vacc_effective` - Respondent's opinion about H1N1 vaccine effectiveness.
1 = Not at all effective; 2 = Not very effective; 3 = Don't know; 4 = Somewhat effective; 5 = Very effective.
17. `opinion_h1n1_risk` - Respondent's opinion about risk of getting sick with H1N1 flu without vaccine.
1 = Very Low; 2 = Somewhat low; 3 = Don't know; 4 = Somewhat high; 5 = Very high.
18. `opinion_h1n1_sick_from_vacc` - Respondent's worry of getting sick from taking H1N1 vaccine.
1 = Not at all worried; 2 = Not very worried; 3 = Don't know; 4 = Somewhat worried; 5 = Very worried.
19. `opinion_seas_vacc_effective` - Respondent's opinion about seasonal flu vaccine effectiveness.
1 = Not at all effective; 2 = Not very effective; 3 = Don't know; 4 = Somewhat effective; 5 = Very effective.
20. `opinion_seas_risk` - Respondent's opinion about risk of getting sick with seasonal flu without vaccine.
1 = Very Low; 2 = Somewhat low; 3 = Don't know; 4 = Somewhat high; 5 = Very high.
21. `opinion_seas_sick_from_vacc` - Respondent's worry of getting sick from taking seasonal flu vaccine.
1 = Not at all worried; 2 = Not very worried; 3 = Don't know; 4 = Somewhat worried; 5 = Very worried.
22. `age_group` - Age group of respondent.
23. `education` - Self-reported education level.
24. `race` - Race of respondent.
25. `sex` - Sex of respondent.
26. `income_poverty` - Household annual income of respondent with respect to 2008 Census poverty thresholds.
27. `marital_status` - Marital status of respondent.
28. `rent_or_own` - Housing situation of respondent.
29. `employment_status` - Employment status of respondent.
30. `hhs_geo_region` - Respondent's residence using a 10-region geographic classification defined by the U.S. Dept. of Health and Human Services. Values are represented as short random character strings.
31. `census_msa` - Respondent's residence within metropolitan statistical areas (MSA) as defined by the U.S. Census.
32. `household_adults` - Number of other adults in household, top-coded to 3.
33. `household_children` - Number of children in household, top-coded to 3.
34. `employment_industry` - Type of industry respondent is employed in. Values are represented as short random character strings.
35. `employment_occupation` - Type of occupation of respondent. Values are represented as short random character strings.

6. Checking if Missing Values in Employment_Occupation Align with the Unemployed

```python
unemployed_status = ['Not in Labor Force', 'Unemployed']

alignment_results = {}

for status in unemployed_status:
    missing_occupation = dataset['employment_occupation'].isna()
    unemployed = dataset['employment_status'] == status
    matching_rows_occupation = sum(missing_occupation & unemployed)

    missing_industry = dataset['employment_industry'].isna()
    matching_rows_industry = sum(missing_industry & unemployed)

    alignment_percentage_occupation = (matching_rows_occupation / len(dataset)) * 100
    alignment_percentage_industry = (matching_rows_industry / len(dataset)) * 100

    alignment_results[status] = {
        'matching_rows': matching_rows_occupation,
        'alignment_percentage': alignment_percentage_occupation,
        'matching_rows_industry': matching_rows_industry,
        'alignment_percentage_industry': alignment_percentage_industry
    }

for status, results in alignment_results.items():
    print(f"Number of rows with missing occupation and '{status}' status: {results['matching_rows']}")
    print(f"Percentage of alignment for '{status}': {results['alignment_percentage']}%")
    print(f"Number of rows with missing industry and '{status}' status: {results['matching_rows_industry']}")
    print(f"Percentage of alignment for '{status}' in industry: {results['alignment_percentage_industry']}%")
    print()
```

- An assumption can be made here that the missing values in the `employment_occupation` and the `employment_industry` columns capture the individuals who were unemployed and therefore did not have an occupation and an industry.
- So, though these columns seem to have incomprehensible information (`they contain random strings`), they happen to tell a story regarding the dataset. 

## Data Preparation / Data Preprocessing
- In this stage, several procedures will be done to ensure that the data is ready for modelling. This section will deal with missing data, scaling of data, handle categorical variables, and conduct train-test split. 

### 2.1: Splitting the Data
- This is done as the first step to ensure there is no data leakage from the test set to the training set.
- Also, splitting the data early ensures that the evaluation of the model will be done on unseen data, and in the end, this will provide a reliable estimate of the model's performance. 

`Note`
- Before splitting the data, the columns `employment_occupation`, `hhs_geo_region`, and `employment_industry` were dropped.
- The reason for dropping these columns was that they contained random strings. No clear meaning could be retrieved from these columns and therefore, they were deemed not useful for the model.

```python
#Dropping the columns 'employment_occupation', 'hhs_region', and 'employment_industry'
dataset = dataset.drop(['employment_occupation', 'hhs_geo_region', 'employment_industry'], axis=1)
dataset.head()
```

- Here, the two target variables are dropped from the `X` variable while the `y` variable it assigned to the seasonal_vaccine target variable.
- As earlier explained, the seasonal_vaccine variable was chosen because there was a significant smaller imbalance between its binary elements.

```python
#Defining X and y
X = dataset.drop(['h1n1_vaccine', 'seasonal_vaccine'], axis=1)
y = dataset['seasonal_vaccine']
```

- A test-size of 0.25 was deemed enough to allow the models to have enough data to train and eventually predict.

```python
#splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### 2.2: Building Pipelines to Replace Null Values in Categorical Columns and Numerical Columns

- Pipelines were deemed crucial in this step because:
1. They help to streamline the workflow. As is seen below, there is a structured and seamless approach to the entire preprocessing steps.
2. They help in reproducibility by encapsulating the entire preprocessing process into a single object. 
3. They help to prevent data leakage, and this is crucial in ensuring reliable estimates from the models. 

```python
#Retrieving the Categorical Columns from the X_train Dataframe

X_train_categorical = X_train.select_dtypes(include='object')

#Retrieving the Numerical Columns from the X_train Dataframe

X_train_numerical = X_train.select_dtypes(exclude='object')
```

```python
#Preprocessing steps for the categorical columns
#Imputing the categorical columns and filling the null values with the word missing and then encoding the columns.
categorical_preprocessing = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent', missing_values=np.nan, fill_value='missing')),
    ('encoder', OrdinalEncoder())
])

#Preprocessing steps for the numerical columns
numerical_preprocessing = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent', missing_values=np.nan))
])
```

- Defining the Column Transformer to Apply the Preprocessing Steps
- The Column Transformer allows the preprocessing steps defined above to be applied to the categorical and numerical columns separately. 

```python
preprocessor = ColumnTransformer([
    ('categorical', categorical_preprocessing, X_train_categorical.columns),
    ('numeric', numerical_preprocessing, X_train_numerical.columns)
])
```

### 3. Feature Selection
- This should be done on the X_train data to avoid data leakage and the risk of overfitting the model to the selected features.
- 25 of the features that were deemed statistically significant and independent of the target variable were selected for modeling.

```python
#Performing feature selection
#Since it is a base model, all features will be used
k = 25

best_features = SelectKBest(chi2, k=k)
```

### 4. Building the Last Pipeline
- This pipeline will combine the previous pipelines and include the feature selection section.

```python
#Combining the previous pipelines together with the feature selection section

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('best_features', best_features)
])
```

## Modeling
- This section will make use of the prepared data to create predictive models. 

### 3.1: Baseline Model: Logistic Regression
- Logistic regression was chosen as the base model because:
1. It is highly interpretable and is a simple model. The results are easier to understand and the relationship between the target and features is easily understood. 
2. It is capable of handling classification problems. 

```python
#Initialize the logistic regression object
logistic_reg = LogisticRegression(max_iter=1000)
```

```python
#Create a pipeline that incorporates the logistic regression model
logistic_pipeline = Pipeline([
    ('pipeline', pipeline),
    ('model', logistic_reg)
])
```

```python
#Fitting the pipeline on the training data
logistic_pipeline.fit(X_train, y_train)
```

#### Evaluating the Base Model

- Checking the Accuracy Score, Cross-Validation, Precision, Recall, F1-Score, and Examining a Confusion Matrix

```python
def eval_model(pipeline, X_train, y_train, X_test, y_test ):
    """This is  a function that calculates the accuracy score, cross-validation, 
    precision, recall, F1-score, and also examines and plots a confusion matrix."""    
    #Cross validation
    cross_val = cross_val_score(pipeline, X_train, y_train, cv=StratifiedKFold(n_splits=10))
    mean_cross_val_score = cross_val.mean()
    print(f'Cross-Validation Mean Accuracy Score: {mean_cross_val_score}')
    print()

    #Checking the accuracy scores of the train and test values
    X_train_predictions = pipeline.predict(X_train)
    X_test_predictions = pipeline.predict(X_test)

    #checking the accuracy of the train values vs the predicted values
    accuracy_train = accuracy_score(y_train, X_train_predictions)
    accuracy_test = accuracy_score(y_test, X_test_predictions)

    #Printing the accuracy scores
    print(f'Accuracy for Train Dataset: {accuracy_train}')
    print(f'Accuracy for Test Dataset: {accuracy_test}')
    print()

    #Checking the Precision, Recall, and F1 score of the base model
    cr = classification_report(y_test, X_test_predictions)
    print(cr)
    print()

    #Instantiate the confusion matrix object
    cm = confusion_matrix(y_test, X_test_predictions)

    #Creating a visual of the matrix
    plt.style.use('default')
    display = ConfusionMatrixDisplay(confusion_matrix=cm)
    display.plot()
    plt.show()

eval_model(logistic_pipeline, X_train, y_train, X_test, y_test)
```

**Observations from the Logistic Regression Model**
1. A mean accuracy score of 0.769 means that the model is correctly predicting the class lables with an accuracy of 76.9%.
2. An accuracy of 0.771 indicates that the model accurately predicts  class labels approximately 77.1% of the time in the train dataset.
3. An accuracy of 0.782 indicates that the model accurately predicts the class labels approximately 78.2% of the time in the test dataset.
4. The model does rather well for both classes, according to the precision, recall, and f1-score values, albeit it might have a little more trouble predicting class 1 than class 0 than they do.
- `Note:` Though the difference in the accuracy score of the model on the training and test dataset is not big, there is some underfitting because the score on the test data is slightly higher. 
Therefore, this justifies the use of other models to try and see if there will be any changes between the two values. 

### 3.2: Random Forest Classifier

- Justification for using random forest classifier
1. It is less sensitive to outliers and any noisy data in the dataset.
2. Reduces overfitting and is believed to provide stable predictions. 
3. Helps to showcase the importance of features in the development of a model. 

```python
#Creating the random forest pipeline
rf_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('best_features', best_features),
    ('scaler', StandardScaler()),
    ('random_forest', RandomForestClassifier())
])

#Defining parameters
rf_param = [{
    'random_forest__n_estimators': [20, 60, 110],
    'random_forest__criterion': ['gini', 'entropy'],
    'random_forest__max_depth': [2,4,6,8],
    'random_forest__min_samples_split': [2]
}]
```

```python
#Instantiating the grid_search
grid_search = GridSearchCV(rf_pipeline, param_grid=rf_param, scoring='accuracy', cv=10)

#Fitting the grid_search
grid_search.fit(X_train, y_train)
```

- Evaluating the Random Forest Model

```python
eval_model(grid_search, X_train, y_train, X_test, y_test)
```

**Observations before Tuning**
1. The cross-validation method yielded a mean accuracy score of 0.774, indicating that the model generally does well in predicting the target variable across various folds of the data. It implies that the model's performance is somewhat predictable and consistent.
2. The model successfully predicts the target variable for about 79.6% of the observations in the training data, according to the accuracy score on the train dataset of 0.796. This shows that the model has picked up on the relationships and patterns found in the training dataset.
3. The test dataset's accuracy score is 0.780, which shows that the model does a good job at extrapolating its predictions to new data. It implies that the model is capable of accurately predicting the target variable for fresh observations and is not overfitting.
4. With regards to precision, recall, and the f1 score, the random forest model shows relatively high precision, recall, and F1-score for both classes, indicating a balanced performance in predicting both positive and negative instances.

- Running the Random Forest Classifier with Best Parameters

```python
#Selecting the best parameters that yield the highest scores

grid_search.best_params_
```

```python
#recreating the model with best parameters
best_params_grid = [{
    'random_forest__criterion': ['gini'],
    'random_forest__max_depth': [8],
    'random_forest__min_samples_split': [2],
    'random_forest__n_estimators': [110]
}]

#fitting the model
best_params_grid_search = GridSearchCV(rf_pipeline, param_grid=best_params_grid, scoring='accuracy', cv=10)
best_params_grid_search.fit(X_train, y_train)
```

```python
#Evaluating the scores after running the model with the best parameters
eval_model(best_params_grid_search, X_train, y_train, X_test, y_test)
```

**Observations**
- Surprisingly, the random forest model with its supposed best estimators happened to procude scores that were lower than before. 
- This necessitated the use of other models.

### 3.3: Gradient Boosting Classifier

- Justification for using the Gradient Boosting Classifier:
1. GBC is able to handle complex relationships between features and still provide strong predictive models.
2. It is believed that GBC, together with its other algorithms like XGBoost have a high predictive power. 
3. Helps to showcase feature importance. 

```python
#Instantiate the pipeline with the gradientboosting classifier
gradient_boosting_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('best_features', best_features),
    ('scaler', StandardScaler()),
    ('gradient', GradientBoostingClassifier())
])

#define the gradient boosting parameters
gradient_params = [{
    'gradient__loss': ['log_loss'],
    'gradient__learning_rate': [0.1, 0.2, 0.3, 0.4],
    'gradient__n_estimators': [20, 60, 120]
}]

#Instantiate the grid search and fit the training dataset on the grid search
gradient_grid_search = GridSearchCV(gradient_boosting_pipeline, param_grid=gradient_params, scoring='accuracy', cv=10)
gradient_grid_search.fit(X_train, y_train)
```

- Evaluating the Gradient Boosting Classifier

```python
#calling the function eval_model to evaluate the gradient boosting classifier
eval_model(gradient_grid_search, X_train, y_train, X_test, y_test)
```

**Observations**
1. The average accuracy across the cross-validation folds appears to be quite high at 77.8%. 
2. The difference between the accuracy score in the train and test datasets shows a slight underfitting problem with the model. 
3. The precision, recall, and F1-scores indicate that the model performs well in classifying both classes (0 and 1), although the performance may be slightly better for class 0.

- Running the Gradient Boost with the Best Parameters
